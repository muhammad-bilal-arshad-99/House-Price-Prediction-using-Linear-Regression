## Data Loading & Preprocessing

```python
from data_preprocessing import load_data, clean_data, prepare_data
from model import train_model
from train import evaluate_model

# Run pipeline
df = load_data()
df = clean_data(df)

X_train, X_test, y_train, y_test, scaler = prepare_data(df)

model = train_model(X_train, y_train)

predictions = evaluate_model(model, X_test, y_test)
```

## Exploratory Data Analysis
This section analyzes relationships between features and house prices.
The scatter plot shows a positive relationship between area and price, indicating that larger houses tend to have higher prices. 
Histograms display the distribution of features, while the heatmap highlights strong correlations, especially between area and price.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot
plt.scatter(df['area'], df['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()

# Histograms
df.hist(figsize=(10,8))
plt.show()

# Correlation heatmap
sns.heatmap(df.corr(), annot=True)
plt.show()
```

## Model Evaluation & Visualization
This section compares actual and predicted values.

The points are close to the diagonal trend, indicating that the model predictions are reasonably accurate.

```python
plt.scatter(y_test, predictions)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted")
plt.show()
```

## Coefficient Analysis 

```python
print(model.coef_)
```

## Model Comparison
This section compares Linear Regression with Decision Tree.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score

# Train Decision Tree
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)

# Predict
dt_pred = dt_model.predict(X_test)

# Compare
print("Linear Regression R2:", r2_score(y_test, predictions))
print("Decision Tree R2:", r2_score(y_test, dt_pred))
```

## Model Interpretation
The model coefficients indicate how each feature affects house price. 
Area has the strongest positive impact, meaning larger houses tend to have higher prices. 
Bedrooms and bathrooms also contribute positively, but with smaller influence compared to area.
The Linear Regression model achieved a higher R² score compared to the Decision Tree model, indicating better generalization and predictive performance.
